```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
import composition_stats as coda
```

```python
# Create data sets
alpha1 = (10, 15, 10)
alpha2 = (50, 10, 30)
alpha3 = (8, 10, 80)
alpha4 = (80, 10, 80)
alpha5 = (500, 900, 7)
X1 = np.random.dirichlet(alpha1, 1000)
X2 = np.random.dirichlet(alpha2, 1200)
X3 = np.random.dirichlet(alpha3, 1000)
#X4 = np.random.dirichlet(alpha4, 500)
#X5 = np.random.dirichlet(alpha5, 500)
X = np.concatenate((X1, X2, X3), axis = 0)
#New transformed data using ilr transformation method
X_new = coda.ilr(X)
```

```python
X_new.shape
```

```python
X.shape
```

```python
model = KMeans(n_clusters=3) #In this example, supppose we already knew the number of clusters
label = model.fit_predict(X_new)
```

```python
#Plot solution
label_plot = np.unique(label)
for i in label_plot:
  plt.rcParams["figure.figsize"] = (10,6)
  plt.scatter(X_new[label == i, 0], X_new[label ==i, 1], label = i)
plt.legend()
plt.show()
```

```python
#Plot solution
label_plot = np.unique(label)
for i in label_plot:
  plt.scatter(X_new[label == i, 0], X_new[label ==i, 1], label = i)
plt.legend()
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], s=300, c='black')
plt.title('Clusters of 5 groups')
plt.show()
```
